### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
'''
Calcular el ingreso total de cada tienda.
'''
import matplotlib.pyplot as plt

lista_tiendas =[tienda,tienda2,tienda3,tienda4]
colores = [ '#33FF57', '#33C1FF', '#FF33A1', '#FF5733','#211C84','#FF71CD', '#6527BE', '#780C28']

def calcular_ingreso(tiendas):
  nombre_tiendas =[]
  ingresos_tiendas = []

  for i, tienda in enumerate(tiendas):
    ingreso_total_tienda = sum(tienda["Precio"])
    print(f"El ingreso total de la tienda {i+1} es: {ingreso_total_tienda:,.2f}$")
    nombre_tiendas.append(f"tienda {i+1}")
    ingresos_tiendas.append(ingreso_total_tienda)

  tienda_mayor_ingresos = max(zip(nombre_tiendas,ingresos_tiendas), key = lambda m: m[1])
  print(f"\nLa {tienda_mayor_ingresos[0]} tiene mayores ingresos de: {tienda_mayor_ingresos[1]:,.2f}$")

  tienda_menor_ingresos = min(zip(nombre_tiendas,ingresos_tiendas), key = lambda m: m[1])
  print(f"La {tienda_menor_ingresos[0]} tiene menores ingresos de: {tienda_menor_ingresos[1]:,.2f}$ \n")

  plt.figure(figsize=(8, 6))
  plt.bar(nombre_tiendas, ingresos_tiendas, color=colores)
  plt.xlabel("Tiendas")
  plt.ylabel("Ingresos Totales ($)")
  plt.title("Comparación de facturación por tienda")
  plt.show()

calcular_ingreso(lista_tiendas)
```

# 2. Ventas por categoría

```python
'''
Agrupar los datos por categoría y contar el número de ventas de cada tipo,
mostrando las categorías más populares de cada tienda.
'''

def ventas_categorias(lista_tiendas):


  for i, tienda in enumerate(lista_tiendas):
    categorias_unicas = set(tienda['Categoría del Producto'])
    ganancias_total_categoria = [[categoria, sum(ganancias for cat,ganancias in zip(list(tienda['Categoría del Producto']),list(tienda['Precio']))
                                  if cat == categoria)] for categoria in categorias_unicas]
    categorias_ventas = {categoria:list(tienda['Categoría del Producto']).count(categoria) for categoria in categorias_unicas}
    categoria_ventas_ganancia = sorted([[cat,ven,monto] for cat,ven in categorias_ventas.items() for categoria, monto in ganancias_total_categoria if categoria == cat], key=lambda m: m[2], reverse=True)
    catgoria_mayor_ingresos = max(categoria_ventas_ganancia, key = lambda c: c[2])

    print("")
    print(f"Tienda {i+1}:")
    total_ventas = 0

    for cat,ven,monto in categoria_ventas_ganancia:
      print(f"{cat}: {ven} ventas, Total facturado: {monto:,.2f}$")
      total_ventas += ven

    print(f"\nLa categoria con mas ganancias es {catgoria_mayor_ingresos[0]} tiene mayores ingresos de: {catgoria_mayor_ingresos[2]:,.2f}$")
    print(F"Total ventas generadas: {total_ventas}")

    categorias_ventas = dict(sorted(categorias_ventas.items(), key=lambda item: item[1], reverse=True))
    plt.figure(figsize=(10, 5))
    plt.bar(categorias_ventas.keys(), categorias_ventas.values(), color=colores)
    plt.xlabel("Categorias")
    plt.ylabel("Numero de ventas Totales")
    plt.title("Comparación de numero de ventas por Categorias")
    plt.xticks(rotation=30, ha='right')
    plt.show()

    print("")
    print("=" * 100)

ventas_categorias(lista_tiendas)
```

# 3. Calificación promedio de la tienda


```python
'''
Debes calcular las calificaciones promedio de los clientes para cada tienda.
El objetivo es conocer la satisfacción del cliente con los productos vendidos.
'''
def calificaciones_tiendas(lista_tiendas):

  i = 1
  t = []
  c = []
  for tienda in lista_tiendas:
    calificaciones_total_tienda = sum(tienda["Calificación"]/len(tienda["Calificación"]))
    print(f"La calificación promedio de la tienda{i} es: {calificaciones_total_tienda:.2f}")
    t.append(f"Tienda {i}")
    c.append(calificaciones_total_tienda)
    i += 1

  calificacion_mayor = max(zip(t,c), key = lambda tc: tc[1])
  calificacion_menor = min(zip(t,c), key = lambda tc: tc[1])
  print(f"\nMas alta {calificacion_mayor[0]} con {calificacion_mayor[1]:.2f}")
  print(f"Mas baja {calificacion_menor[0]} con {calificacion_menor[1]:.2f} \n")
  #print(c)

  def promedio_autopct(c, p):
      p = c * sum(p) / 100
      return f'{p:.2f}'

  plt.figure(figsize=(6, 6))
  plt.pie(c, labels=t, autopct=lambda cat: promedio_autopct(cat, c),startangle=90, colors=colores)
  plt.title('Clasificaciones Promedio por Tienda')
  plt.axis('equal')
  plt.show()
  plt.show()


calificaciones_tiendas(lista_tiendas)
```

# 4. Productos más y menos vendidos

```python
'''
Debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda.
Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.
'''
def mayor_menor_ventas(tiendas):
  i = 1

  for tienda in tiendas:
    productos = set(tienda['Producto'])
    ventas_por_producto = [(producto, list(tienda['Producto']).count(producto)) for producto in set(tienda['Producto'])]

    numero_mayor_de_ventas = max([mas_vendido[1] for mas_vendido in ventas_por_producto])
    productos_con_mayor_ventas = [(producto[0], producto[1]) for producto in ventas_por_producto if producto[1] == numero_mayor_de_ventas]

    numero_menor_de_ventas = min([mas_vendido[1] for mas_vendido in ventas_por_producto])
    productos_con_menor_ventas = [(producto[0], producto[1]) for producto in ventas_por_producto if producto[1] == numero_menor_de_ventas]

    print("=" * 100)
    print(f'''\nTienda {i}: \n\tProductos con mayor ventas:''')
    for p, v in productos_con_mayor_ventas:
      print(f"\t {p} con {v} ventas")

    print(f'''\n\tProductos con menor ventas:''')
    for p, v in productos_con_menor_ventas:
      print(f'''\t {p} con {v} ventas''')
    pro = [producto[0] for producto in productos_con_mayor_ventas+productos_con_menor_ventas]
    ven = [ventas[1] for ventas in productos_con_mayor_ventas+productos_con_menor_ventas]

    plt.figure(figsize=(5, 3))
    plt.bar(pro, ven, color=colores)
    plt.xlabel("Categorias")
    plt.ylabel("Numero de ventas Totales")
    plt.title("Comparación de numero de ventas por Categorias")
    plt.xticks(rotation=30, ha='right')
    plt.show()
    i += 1

mayor_menor_ventas(lista_tiendas)
```

# 5. Envío promedio por tienda

```python
'''
Debes calcular el costo de envío promedio para cada tienda.
El objetivo es comprender cuánto se gasta, en promedio, en el Costo de envío de cada tienda.
'''
def costo_envio(tiendas):
  i = 1
  t = []
  e = []
  for tienda in tiendas:
    costo_envio_total_tienda = sum(tienda["Costo de envío"]/len(tienda["Costo de envío"]))
    print(f"El costo de envío promedio de la tienda {i} es: {costo_envio_total_tienda:,.2f}$")
    t.append(f"Tienda {i}")
    e.append(costo_envio_total_tienda)
    i += 1


  tienda_mayor_gasto = max(zip(t,e), key = lambda te: te[1])
  print(f"\nEl coste de envio mas alto es en la {tienda_mayor_gasto[0]} con {tienda_mayor_gasto[1]:,.2f}$ \n")
  plt.figure(figsize=(5, 3))
  plt.bar(t,e, color=colores)
  plt.xlabel("Tiendas")
  plt.ylabel("Promedio de envíos totales")
  plt.title("Comparación de envíos promedio por tienda")
  plt.xticks(rotation=30, ha='right')
  plt.show()


costo_envio(lista_tiendas)
```
